```python
def daily_sma():
  symbol = 'BTCUSDT'
  num_bars = 70 # Number of rows that we get, each one is of format: [Timestamp, Open, High, Low, Close, Volume]
  timeframe = '1d' # The timestamp difference of each row  
  bars = exchange.fetch_ohlcv(symbol, timeframe = timeframe, limit = num_bars) 

  df_d = pd.DataFrame(bars, columns = ['timestamp', 'open', 'high', 'low', 'close', 'volume'])
  df_d['timestamp'] = pd.to_datetime(df_d['timestamp'], unit = 'ms') # Convert timestamp to proper date/time format instead of milliseconds
  df_d['sma20_d'] = df_d.close.rolling(20).mean() # Each row in the 'sma20_d' column contains the average of the last 20 closing values, the first 19 are null which we are dropping
  df_d = df_d.dropna()

  bid = ask_bid()[1]
  df_d.loc[df_d['sma20_d'] > bid, 'sig'] = 'SELL' # sma > bid = SELL
  df_d.loc[df_d['sma20_d'] < bid, 'sig'] = 'BUY' # sma < bid = BUY

  return df_d
```
