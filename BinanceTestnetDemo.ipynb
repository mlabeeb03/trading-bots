```python
import ccxt
import pandas as pd
import json
```

```python
def StoreInFile(object):
  with open('Log.txt', 'a') as Log:
     Log.write(json.dumps(object))
     Log.write('\n')
```

```python
def ReadFile():
  Lines = []
  with open('Log.txt', 'r') as Log:
    for i in Log.readlines():
      Lines.append(json.loads(i))
  for i in Lines:
    print(i)
```

```python
def BuyShort(symbol, amount):
  try:
    binance.set_position_mode(hedged=True)
  except:
    pass
  symbol = symbol
  type = 'market'
  side = 'sell'
  amount = amount
  price = None
  params = {'positionSide': 'SHORT'}
  order = binance.create_order(symbol, type, side, amount, price, params)
  print(order)
  StoreInFile(order)
```
